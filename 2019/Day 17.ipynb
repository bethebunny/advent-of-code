```python
class OutputEnv(StdioEnv):
    def reset(self):
        self.outputs = []
        
    def output(self, value):
        self.outputs.append(value)
```

```python
ctx = interpreter.evaluate(program, env=OutputEnv())
```

```python
print(''.join(chr(i) for i in ctx.env.outputs))
```

```python
map = [list(row) for row in ''.join(chr(i) for i in ctx.env.outputs).strip().split("\n")]
```

```python
mapd = {(x, y): map[y][x] for y in range(len(map)) for x in range(len(map[0]))}
```

```python
def neighbors(point):
    x, y = point
    for i in -1, 1:
        yield (x + i), y
        yield x, (y + i)
```

```python
def is_intersection(mapd, point):
    return mapd[point] == '#' and all(mapd.get(n, None) == '#' for n in neighbors(point))
```

```python
[p for p in mapd if is_intersection(mapd, p)]
```

```python
sum(p[0] * p[1] for p in mapd if is_intersection(mapd, p))
```

```python
list(neighbors((11, 28)))
```

```python
def print_mapd_with_intersections(map, mapd):
    for y, row in enumerate(map):
        row_str = ''
        for x, v in enumerate(row):
            if is_intersection(mapd, (x, y)):
                row_str += 'O'
            else:
                row_str += v
        print(row_str)
```

```python
print_mapd_with_intersections(map, mapd)
```
